```python
pydataset.data("iris", show_doc=True)
```

```python
iris = pydataset.data("iris")
```

```python
plt.scatter(iris["Sepal.Length"], iris["Sepal.Width"], sizes=20*iris["Petal.Length"])  
plt.show()
```

```python
def especie_color(x):
    if x == "setosa":
        return 0
    elif x == "versicolor":
        return 1
    return 2

iris["SpeciesNumber"] = iris["Species"].apply(especie_color)

plt.scatter(
    iris["Sepal.Length"], iris["Sepal.Width"], sizes = 15*iris["Petal.Length"],
    c = iris["SpeciesNumber"], cmap="inferno", alpha=0.4 )
plt.show()
```
